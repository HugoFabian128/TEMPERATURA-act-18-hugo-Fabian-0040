## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos relaciones lineales basadas en los puntos de congelación (0°C o 32°F) y ebullición (100°C o 212°F) del agua.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(También se puede usar $1.8$ en lugar de la fracción $9/5$)*.

* **De Fahrenheit a Celsius:**
    $$°C = \frac{(°F - 32) \times 5}{9}$$

## 2. Ejemplos Resueltos Paso a Paso

### Caso A: Convertir 25°C a °F (Temperatura ambiente)
1.  **Sustituir:** Ponemos el 25 en la fórmula: $°F = (25 \times \frac{9}{5}) + 32$.
2.  **Multiplicar:** $25 \times 9 = 225$. Luego $225 / 5 = 45$.
3.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** **77°F**.

### Caso B: Convertir 98.6°F a °C (Temperatura corporal promedio)
1.  **Sustituir:** $°C = (98.6 - 32) \times \frac{5}{9}$.
2.  **Restar primero:** $98.6 - 32 = 66.6$.
3.  **Multiplicar y dividir:** $66.6 \times 5 = 333$. Luego $333 / 9 = 37$.
* **Resultado:** **37°C**.

## 3. Código en Python

Este programa es interactivo y sigue exactamente la estructura de funciones que solicitaste.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú y gestiona la lógica principal"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`while True`:** Crea un bucle infinito para que el menú aparezca una y otra vez hasta que el usuario decida salir.
* **`float(input(...))`:** Permite que el usuario ingrese números con decimales.
* **`f"{variable:.2f}"`:** Es una forma de dar formato al texto para que solo muestre 2 decimales, evitando números muy largos en pantalla.
* **`break`:** Es la instrucción clave para romper el bucle y terminar la ejecución cuando se elige la opción 3.
